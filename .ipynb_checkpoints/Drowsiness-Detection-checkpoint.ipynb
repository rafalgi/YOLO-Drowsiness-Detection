# Instal and import dependence

```python
import torch
import matplotlib.pyplot as plt
import cv2
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

# Load Model

```python
model = torch.hub.load('ultralytics/yolov5' , 'yolov5s')
```

```python
model
```

# Make Detections

```python
img = cv2.imread('pies.jpg')
```

```python
result = model(img)
result.print()
```

```python
%matplotlib inline 
plt.imshow(np.squeeze(result.render()))
plt.show()
```

```python
# Wczytanie obrazów
images = [
    cv2.imread('kot.jpg'),
    cv2.imread('kot1.jpg'),
    cv2.imread('kot3.jpg'),
    cv2.imread('psy1.jpg')
]

# Utworzenie figury i osi (subplotów) w siatce 2x2
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Iteracja przez osie i obrazy, aby je wyświetlić
for ax, img in zip(axes.flat, images):
    ax.imshow(img)

# Wyświetlenie wykresu
plt.show()
```

```python
# Przetwarzanie obrazów za pomocą modelu YOLOv5
results = [model(img) for img in images]

# Utworzenie figury i osi (subplotów) w siatce 2x2
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Iteracja przez osie i przetworzone obrazy, aby je wyświetlić
for ax, result in zip(axes.flat, results):
    # Renderowanie obrazu z zaznaczonymi obiektami
    img = np.squeeze(result.render())
    
    ax.imshow(img)  # Wyświetlenie obrazu (w formacie BGR)
    ax.axis('off')  # Wyłączenie osi dla estetyki

# Wyświetlenie całości
plt.show()
```

# Real time Detections

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

# Train from scratch

```python
import uuid
import os
import time
```

```python
IMAGES_PATH = os.path.join('data' , 'images')
labels  = ['awake' , 'drowsy']
number_imgs = 20
```

```python
cap = cv2.VideoCapture(0)
# Loop through labels
for label in labels:
    print(f'Collecting images for {label}')
    time.sleep(5)
    
    # Loop through images
    for image_num in range(number_imgs):
        print(f'Collecting images for {label} , image number{image_num}' )
        ret , frame = cap.read()

        imgname  = os.path.join(IMAGES_PATH , label +"."+str(uuid.uuid1()+'.jpg')) # Gdzie mają być zapisywany , nazwa obrazu
        cv2.imwrite(imgname ,frame)
        cv2.imshowo('Image Collection' , 'Frame')
        time.sleep(2)
```
